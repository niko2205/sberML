```python
import pandas as pd
import numbers
import math
```

```python
raw.columns = ['DealID', 'CreditSumm', 'Rating', 'CreditHistory', 'Verifycation', 'VoteCI', 'VoteCM', 'PledgeSumm', 'Decision']
raw.head()
```

```python
def replace_letters(x):
    if isinstance(x, numbers.Number):
        return x
    return x.replace('ё', 'е')
```

```python
raw['CreditHistory']=raw['CreditHistory'].apply(replace_letters)
raw['Decision']=raw['Decision'].apply(replace_letters)
raw.head()
```

```python
def decision_to_digit(x):
    if x == 'Одобрен':
        return 1
    return 0
```

```python
raw['Decision']=raw['Decision'].apply(decision_to_digit)
raw.head()
```

```python
def is_pledge_exists(x):
    if math.isnan(x):
        return 0
    return 1
```

```python
raw['IsPledgeExists']=raw['PledgeSumm'].apply(is_pledge_exists)
raw.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = raw.drop(['DealID', 'Decision'], axis=1)
y = raw['Decision']
print('X length = '+ str(len(X)))
print('y length = '+ str(len(y)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
print('X_train length = '+ str(len(X_train)))
print('X_test length = '+ str(len(X_test)))
print('y_train length = '+ str(len(y_train)))
print('y_test length = '+ str(len(y_test)))
```
